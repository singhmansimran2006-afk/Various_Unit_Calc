Celcius to Fahrenheit

```python
C = float(input("Enter the Temperature"))
F = float ((C * 1.8) + 32)         

print ("Your Temperature in F is :", F)
```

Celcius to Fahrenheit

```python
C = float(input("Enter your Temperature"))
K = float( C + 273.15 )
print ("Your Temperature in K is :", K)
```

MilliMetres to Inches

```python
mm = float(input("Enter your value"))
In = float( mm/25.4 )
print("Your value in Inches is : ", In )
```

Kilometres to Metres

```python
km = float(input("Enter your distance"))
m = float( km * 1000 )
print ("Your distance in m is : ", m)           
```

Days to Seconds

```python
days = float(input("Enter your value in days"))
sec = float(days * 86400)
print ("Your value in seconds is : ", sec)
```
